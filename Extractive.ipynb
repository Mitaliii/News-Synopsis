```python
import nltk
import numpy as np
import re
import os
```

```python
class Preprocessing(object):

    def __init__(self):
        # self.factory = StemmerFactory()
        # self.stemmer = self.factory.create_stemmer()
        self.stopwords = [line.rstrip('\n\r') for line in open(os.path.join(os.getcwd(), 'C:\\Users\\User\\Documents\\Model\\stopwords.txt'))]

    def casefolding(self, sentence):
        sentence = sentence.lower()
        sentence = re.sub(r'[^a-z]', ' ', re.sub("â€™", '', sentence))
        return sentence

    def tokenization(self, sentence):
        return sentence.split()

    def stopword_removal(self, token):
        temp = []
        for i in range(len(token)):
            if token[i] not in self.stopwords:
                temp.append(token[i])
        return temp
```

```python
import nltk
def fit(paragraph):
    def word_freq(data):
        w = []
        for sentence in data:
            for words in sentence:
                w.append(words)
        bag = list(set(w))
        res = {}
        for word in bag:
            res[word] = w.count(word)
        return res

    pre = Preprocessing()

    sentence_list =  nltk.sent_tokenize(paragraph)
    data = []
    for i in range(len(sentence_list)):
        data.append(pre.stopword_removal(pre.tokenization(pre.casefolding(sentence_list[i]))))
    data = (list(filter(None, data)))

    wordfreq = word_freq(data)

    ranking = []
    for words in data:
        temp = 0
        for word in words:
            temp += wordfreq[word]
        ranking.append(temp)

    sort_list = sorted(range(len(ranking)), key=ranking.__getitem__, reverse=True)
    n = 5
    sentence = ''
    for i in range(n):
        sentence += '{} '.format(sentence_list[sort_list[i]])
    return sentence
```

```python
def get_summary():
      url = utext.get('1.0',"end").strip()
      article = Article(url)
      article.download()
      article.parse()
      article.nlp()
      text = article.text
      summarize = fit(text)
      return summarize
```
